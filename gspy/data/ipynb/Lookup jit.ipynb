```python
import numpy as np
import pandas as pd
# from numba.decorators import jit
from numba import jit, autojit, double, f8, i8
from scipy.io import loadmat
```

```python
data = loadmat('lookup.mat')
tabvals = data['tab'].squeeze()
endadj = float(data['endadj'])
x = data['x'].squeeze()
```

```python
def lookup(tabvals, x, endadj):
    """
    Just a function that matches lookup.c or lookup.m from compecon

    This is correct, at least for init_coefs case
    """
    n = np.prod(x.shape)
    m = tabvals.size
    if endadj >= 2:
        m = m - np.where(tabvals == tabvals[-1])[0].size

    temp_series = pd.Series(np.append(tabvals[:m], x))
    temp_series.sort()
    ind = temp_series.index.values
    temp = np.where(ind > m - 1)[0]
    j = ind[temp] - m
    ind = (temp - range(1, n + 1)).reshape(x.shape)
    ind = ind[j]
    if endadj == 1 or endadj == 2:
        ind[ind == 0] = np.where(tabvals == tabvals[0])[0].size
    return ind
```

```python
fast_look = jit('f8[:,:](f8[:,:],f8[:,:],i8)')(lookup)
```

```python
flex_lookup = autojit(lookup)
```

```python
%timeit lookup(tabvals, x, endadj)
```
